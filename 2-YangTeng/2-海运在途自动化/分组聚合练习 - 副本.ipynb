```python
source = r'D:\Asher\运营\运营交接\20230818\【新老品在途管控表】US-ebay 20230818.xlsx'
```

```python
import pandas as pd
```

```python
data = pd.read_excel(source,sheet_name=0).fillna('')

data
```

```python
data['备注'] = '上架时间:'
data['到港时间'] = '到港时间:'
data['实际上架时间'] = ''
```

```python
group_data = data[['跟踪号']].drop_duplicates(keep='first').reset_index(drop=True)

group_data
```

```python
group_data['编号'] = [int(i + 1) for i in range(len(group_data))]
```

```python
group_data
```

```python
import numpy as np
```

```python
data[['跟踪号','单号','预报数量','SKU']][data['跟踪号']==group_data.loc[0,'跟踪号']]
```

```python
pivot_table = pd.DataFrame()
```

```python
for i in range(len(group_data)):
    print(group_data.loc[i,'跟踪号'])
    explode_data = data[['跟踪号','单号','预报数量','SKU']][data['跟踪号']==group_data.loc[i,'跟踪号']]
    explode_pivot_table = explode_data.pivot_table(index=['跟踪号','单号'], values=['预报数量','SKU'], aggfunc={'预报数量':np.sum,'SKU':'count'}, margins=True, margins_name='Total')
    pivot_table = pd.concat([pivot_table,explode_pivot_table])
```

```python
pivot_table
```

```python
pivot_table.reset_index()
```

```python
num_data = data[['跟踪号','单号','预报总箱数']].drop_duplicates(subset=['单号'],keep='first')

num_data.reset_index(drop=True,inplace=True)

num_data
```

```python
num_pivot_table = pd.DataFrame()
```

```python
for i in range(len(group_data)):
    print(group_data.loc[i,'跟踪号'])
    num_explode_data = num_data[num_data['跟踪号']==group_data.loc[i,'跟踪号']]
    num_explode_pivot_table = num_explode_data.pivot_table(index=['跟踪号','单号'], values=['预报总箱数'], aggfunc={'预报总箱数':np.sum}, margins=True, margins_name='Total')
    num_pivot_table = pd.concat([num_pivot_table,num_explode_pivot_table])
```

```python
num_pivot_table
```

```python
all_pivot = pd.concat([pivot_table,num_pivot_table],axis=1)

all_pivot
```

```python
left_pivot = data[['发货仓库','物流商','发货柜型','发货类型','跟踪号','发货时间','预计上架时间','预计时效','入库单状态','备注','预计上架时间-更新','到港时间','实际上架时间']].drop_duplicates(subset=['跟踪号'],keep='first')

left_pivot.reset_index(drop=True,inplace=True)

left_pivot
```

```python
left_pivot = pd.merge(group_data,left_pivot,on='跟踪号',how='left')

left_pivot
```

```python
reset_all_pivot = all_pivot.reset_index()

reset_all_pivot
```

```python
Total_sum_up = reset_all_pivot[reset_all_pivot['跟踪号']=='Total'].pivot_table(index=['跟踪号'], values=['SKU','预报数量','预报总箱数'], aggfunc={'SKU':np.sum,'预报数量':np.sum,'预报总箱数':np.sum}, margins=True, margins_name='All_Total')

Total_sum_up
```

```python
Total_sum_up.reset_index().iloc[[1]]
```

```python
total_pivot = pd.concat([reset_all_pivot,Total_sum_up.reset_index().iloc[[1]]])

total_pivot
```

```python
left_pivot
```

```python
pd.merge(left_pivot,total_pivot.reset_index(drop=True),on='跟踪号',how='right')
```

```python
final_result = pd.merge(left_pivot,total_pivot.reset_index(drop=True),on='跟踪号',how='right').fillna('').set_index(['跟踪号','编号','发货仓库','物流商',
                                                                            '发货柜型','发货类型','发货时间','预计上架时间','预计时效','入库单状态','备注','预计上架时间-更新','到港时间','实际上架时间','单号'])

final_result
```
